### Project Part 1: Improve Data Visualization Skills

#### 1. Three Principles of effective data communication:
1) Your data should have an order, someone views elements of a graph in an order, your visualization should have an order that makes sense to a viewer.
2) Hierachry - Viewers have a hierarchy of things that they look at, utilize brighter and darker colors to tell your story
3) Clarity - Doing simpler is better. Don't overcomplicate your graphs with too much material otherwise the viewer isn't going to be able to quickly interpret your graph.

#### 2. The basics of Visual Perception: 
- Order: The order that someone views elements of a graph
- Hierarchy: Such as viewing brighter colors over darker colors
- Clarity: Simpler is better, people can only remember 2 or 3 things at a time
- Relationships: Viewers tend to view things in relationships, they will form relationships with things that appear in groups
- Convention: There is a generally understood "convention" i.e. green means good, red means bad

These concepts will help create more clear visualizations. I'd be able to "tell the story" with the data more effectively and efficiently to a client.

#### 3. Choose Appropriate graphs for the Data
- Comparison between values: Bar/column chart
- Comparison to the whole: Stacked column or treemap column
- Change over time: line graph
- Ranking data: Bar, or slope
- Correlation: scatter plot, bubble plot
- Geographical Charts: geographical maps
- Measuring a target: Grouped columns
- Showing Outliers: Box and whisker plot

Choosing the right graph seems essential. It's the best way to show the data in the quickest way that makes sense.

#### 5. Decluttering Visualizations

- What is visualization clutter? : This is when you have too much content on your graph, i.e. too many labels, too many numbers, too many lines, too many colors etc. It takes away from the overall message of the visualization
- What are the main components of a graph: The x axis, y axis, legend, titles, and content
- What are three techniques you learned to make data visualizations more clear? : Use colors to show grouping with data. Use light and dark backgrounds so draw attention to your message. A lot of times less is more.
- How can the use of color affect the way your visualizations are understood? : Too many colors can add to overall clutter. It can be used to be decieving by using convention (red and green) colors inappropriately.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('salary.csv')
df_n = pd.read_csv('salary_numeric.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['gender', 'education', 'industry', 'industry_experience', 'salary']]
```

```python
df_gender = df.groupby('gender')
df_gender = df.reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='salary', hue='gender', palette='mako', bins=18, kde=True)

plt.xlabel('Salary', fontsize=20)
plt.ylabel('Number of Managers', fontsize=20)
plt.title('Number of Male vs Female Managers', fontsize=25)

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_gender, x='education', y='salary', hue='gender', palette='mako')

plt.xlabel('Education', fontsize=20)
plt.ylabel('Salary', fontsize=20)
plt.title('Male and Female Salaries by Education Level', fontsize=25)

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_gender, x='industry_experience', y='salary', hue='gender', palette='mako', ci=None, linewidth=3)

plt.xlabel('Yrs Industry Experience', fontsize=20)
plt.ylabel('Salary', fontsize=20)
plt.title('Male and Female Salaries by Industry Experience', fontsize=25)
plt.legend(loc='lower right')

plt.show()
```
